```python
def rule_30(state):
    return (state >> 1) ^ (state | state << 1)

cells_per_row = 10000
start_state = 1 << (cells_per_row // 2)
num_iterations = 120

state = start_state
mid = []
for _ in range(num_iterations):
    print(''.join('O' if state >> j & 1 else '.' for j in range(cells_per_row - 1, -1, -1)))
    mid.append(state >> (cells_per_row // 2) & 1)
    state = rule_30(state)
```
